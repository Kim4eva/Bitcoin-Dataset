```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

decomposition = seasonal_decompose(df['Close'], model='additive', period=365)


plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()
```

```python
df['Volatility_30'] = df['Close'].rolling(window=30).std()
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Volatility_30'], label='30-Day Rolling Volatility', color='purple')
plt.title('Bitcoin 30-Day Volatility Over Time')
plt.xlabel('Date')
plt.ylabel('Volatility (Standard Deviation)')
plt.legend()
plt.show()
```

```python
# Calculate correlation matrix
correlation_matrix = df[['Close', 'Volume', 'SMA_30', 'SMA_90']].corr()

print(correlation_matrix)
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model
model = ARIMA(df['Close'], order=(5, 1, 0))  # (p, d, q) parameters
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=30)  # Forecast for the next 30 days
print(forecast)
```

```python
df = pd.read_csv('C:/Users/surface/Downloads/BTC-USD(1).csv')
```

```python
df['SMA_30'] = df['Close'].rolling(window=30).mean()  # 30-day SMA
df['SMA_90'] = df['Close'].rolling(window=90).mean()  # 90-day SMA

# Plot the closing prices with SMAs
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Closing Price', color='grey')  # Set Closing Price to grey
plt.plot(df['Date'], df['SMA_30'], label='30-Day SMA', color='#FF6F61')  # Set 30-Day SMA to pinkish red
plt.plot(df['Date'], df['SMA_90'], label='90-Day SMA', color='red')  # Set 90-Day SMA to red

# Customize the plot
plt.title('Bitcoin Closing Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price($)')

# Modify legend to remove the box and set location
plt.legend(frameon=False, loc='upper left')  # Remove the box around the legend and position it better
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Decompose the time series (multiplicative model)
# Adjust 'period' to a suitable value based on the frequency of your data.
decomposition = seasonal_decompose(df['Close'], model='multiplicative', period=365)

# Plot the decomposed components
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()
```
